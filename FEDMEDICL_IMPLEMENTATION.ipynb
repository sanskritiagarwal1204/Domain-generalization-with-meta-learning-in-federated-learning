### Fitzpatrick

```python
import os
import sys
import json
from os.path import dirname as up

# Get the directory where this script resides using __file__
current_dir = os.path.dirname(os.path.abspath(__file__))

# Compute the directory that should contain preprocess_helper.py
# Assuming preprocess_helper.py is in the parent folder of this file (i.e., in ../)
helper_dir = os.path.abspath(os.path.join(current_dir, os.pardir))
if helper_dir not in sys.path:
    sys.path.insert(0, helper_dir)

# Now import from preprocess_helper
from preprocess_helper import preprocess_metadata, split_train_test, save_split

def preprocess_fitzpatrick17k():
    # Compute the project root relative to this script.
    # For our assumed structure: from preprocessed_dataset go up two levels to FedMedICL.
    project_root = os.path.abspath(os.path.join(current_dir, os.pardir, os.pardir))
    config_path = os.path.join(project_root, "configs", "datasets.json")

    with open(config_path, "r") as f:
        dataset_config = json.load(f)

    dataset_path = dataset_config["fitzpatrick17k"]["path_to_dataset"]
    metadata_path = os.path.join(dataset_path, "fitzpatrick17k.csv")
    selected_columns = ["md5hash", "fitzpatrick_scale", "three_partition_label", "Path"]
    image_id_col = "md5hash"
    image_folder = "images"
    target_column = "three_partition_label"
    destination_path = os.path.join(dataset_path, "split")

    metadata = preprocess_metadata(
        dataset_path, metadata_path, selected_columns, image_id_col, image_folder, target_column
    )

    # Remove rows where fitzpatrick_scale is -1 (i.e. null)
    metadata = metadata[metadata["fitzpatrick_scale"] != -1]

    # Define skin_type (0-indexed)
    metadata["skin_type"] = metadata["fitzpatrick_scale"] - 1

    # Convert skin_type to binary (0 if skin_type <= 2, else 1)
    skin_lists = metadata["skin_type"].values.tolist()
    metadata["skin_binary"] = [0 if x <= 2 else 1 for x in skin_lists]
    metadata.drop(["fitzpatrick_scale"], axis=1, inplace=True)

    print("Final Columns:", metadata.columns.values)
    train_split, other_split = split_train_test(metadata, image_id_col, test_size=0.2)
    val_split, test_split = split_train_test(other_split, image_id_col, test_size=0.5)
    save_split(
        train_split, destination_path, "train.csv", config_path, "fitzpatrick17k", "train_meta_path"
    )
    save_split(
        val_split, destination_path, "val.csv", config_path, "fitzpatrick17k", "val_meta_path"
    )
    save_split(
        test_split, destination_path, "test.csv", config_path, "fitzpatrick17k", "test_meta_path"
    )

if __name__ == "__main__":
    preprocess_fitzpatrick17k()
```

### OL3I

```python
import sys
[sys.path.append(i) for i in [".", ".."]]
# Adjust the following import if your helper is in a subfolder; otherwise, use as-is.
sys.path.insert(0, '/content/FedMedICLFolder/FedMedICL')
from preprocessing.preprocess_helper import preprocess_metadata, process_age, process_sex, save_split

import pandas as pd
import json
import os
from tqdm import tqdm
import h5py
from pathlib import Path
import numpy as np
from PIL import Image

# Set pandas chained assignment warning off
pd.options.mode.chained_assignment = None  # default='warn'

def preprocess_data(data):
    """
    Preprocesses the input data (image) to fit the right shape and pixel range.
    Args:
        data (numpy.ndarray): The input data array with shape (512, 512).
    Returns:
        numpy.ndarray: The preprocessed data array with shape (3, 512, 512).
    """
    # Clip the data to the range -1000 to 1000
    clipped_data = np.clip(data, -1000, 1000)
    # Reshape to 3D (1, 512, 512) and repeat to get 3 channels
    reshaped_image = clipped_data.reshape(1, 512, 512)
    reshaped_data = np.repeat(reshaped_image, 3, axis=0)
    # Scale to range (0, 1)
    normalized_data = (reshaped_data - (-1000)) / (1000 - (-1000))
    # Scale to range (0, 255)
    scaled_data = normalized_data * 255
    # Round and convert to uint8
    rounded_data = np.round(scaled_data).astype(np.uint8)
    return rounded_data

def save_as_jpg(data, filename):
    """
    Saves the input data array as a JPEG image.
    Args:
        data (numpy.ndarray): The input data array with shape (3, 512, 512).
        filename (str): The filename to save the image as.
    """
    image = Image.fromarray(np.transpose(data, (1, 2, 0)))
    image.save(filename, "JPEG")

def prerpocess_ol3i_images(metadata_path, hdf5_path, image_path, image_id_col):
    """
    Preprocess OL3I images based on metadata and save them as JPEG files.
    Args:
        metadata_path (str): Path to the metadata file.
        hdf5_path (str): Path to the HDF5 data file.
        image_path (str): The path to store the images.
        image_id_col (str): The name of the column containing the image IDs.
    """
    # Create the image_path directory if it does not exist
    Path(image_path).mkdir(parents=True, exist_ok=True)
    # Load the metadata CSV
    metadata = pd.read_csv(metadata_path)
    print("\nThe preprocessed metadata includes the following columns:\n", metadata.columns.values)
    # Open the HDF5 file
    with h5py.File(hdf5_path, "r") as hdf_file:
        for i in tqdm(range(len(metadata))):
            anon_id = metadata.iloc[i][image_id_col]
            data = hdf_file[anon_id][()]
            image = preprocess_data(data)
            save_as_jpg(image, f"{image_path}/{anon_id}.jpg")
    print(f"\nProcessed and stored {i} images.\n")

def preprocess_OL3I():
    # Use an absolute path for the configuration file.
    config_path = '/content/FedMedICLFolder/FedMedICL/configs/datasets.json'
    with open(config_path, "r") as f:
        dataset_config = json.load(f)
    # Get dataset path from JSON for OL3I
    dataset_path = dataset_config["OL3I"]["path_to_dataset"]
    # Paths for metadata and HDF5 file
    metadata_path = os.path.join(dataset_path, "clinical_data.csv")
    hdf5_path = os.path.join(dataset_path, "l3_slices.h5")
    # Define the columns we want to use; adjust as needed.
    selected_columns = ["anon_id", "bmi", "smoker", "age", "sex", "label_1y", "set_1y", "Path"]
    image_id_col = "anon_id"
    # Specify the folder where images will be saved
    image_folder = "OL3I_images"
    image_path = os.path.join(dataset_path, image_folder)
    target_column = "label_1y"
    destination_path = os.path.join(dataset_path, "split")

    # If the image folder doesn't exist, extract images from the HDF5 file.
    if not os.path.isdir(image_path):
        prerpocess_ol3i_images(metadata_path, hdf5_path, image_path, image_id_col)

    # Preprocess metadata (this function should handle reading the CSV and applying necessary transformations)
    metadata = preprocess_metadata(dataset_path, metadata_path, selected_columns, image_id_col, image_folder, target_column)
    metadata = process_sex(metadata, "sex")
    metadata = process_age(metadata, "age")
    print("Final Columns:", metadata.columns.values)
    # For OL3I, use the 'set_1y' column to split the data
    train_split = metadata[metadata["set_1y"] == "train"]
    val_split = metadata[metadata["set_1y"] == "val"]
    test_split = metadata[metadata["set_1y"] == "test"]
    save_split(train_split, destination_path, "train.csv", config_path, "OL3I", "train_meta_path")
    save_split(val_split, destination_path, "val.csv", config_path, "OL3I", "val_meta_path")
    save_split(test_split, destination_path, "test.csv", config_path, "OL3I", "test_meta_path")
    print("OL3I preprocessing complete. Splits saved in:", destination_path)

if __name__ == "__main__":
    preprocess_OL3I()
```

### PAPILA

```python
import sys
[sys.path.append(i) for i in [".", ".."]]
sys.path.insert(0, '/content/FedMedICLFolder/FedMedICL')

from preprocessing.preprocess_helper import (
    preprocess_metadata,
    process_age,
    process_sex,
    split_train_test,
    save_split,
)
import pandas as pd
import json
import os

# Use an absolute path for the configuration file.
config_path = '/content/FedMedICLFolder/FedMedICL/configs/datasets.json'

def combine_metadatas(dataset_path, selected_columns, image_folder, save_path):
    """
    Combines the metadata from the OD and OS Excel files and saves the selected columns to a new CSV file.
    """
    od_meta = pd.read_excel(os.path.join(dataset_path, "ClinicalData", "patient_data_od.xlsx"))
    os_meta = pd.read_excel(os.path.join(dataset_path, "ClinicalData", "patient_data_os.xlsx"))

    # Assume "Unnamed: 0" is used as the identifier column.
    os_meta["Path"] = [image_folder + "/RET" + str(x)[1:] + "OS.jpg" for x in os_meta["Unnamed: 0"].values]
    od_meta["Path"] = [image_folder + "/RET" + str(x)[1:] + "OD.jpg" for x in od_meta["Unnamed: 0"].values]

    # Combine the OD and OS metadata into a single DataFrame
    meta_all = pd.concat([od_meta, os_meta])

    # Select the relevant columns and save the metadata to a new CSV file
    meta_all = meta_all[selected_columns]
    meta_all.to_csv(save_path, index=False)
    print("Combined metadata saved to", save_path)

def preprocess_PAPILA():
    # Use the absolute path for the configuration file.
    with open(config_path, "r") as f:
        dataset_config = json.load(f)

    # Use the PAPILA dataset path from your JSON
    dataset_path = dataset_config["PAPILA"]["path_to_dataset"]
    metadata_path = os.path.join(dataset_path, "ClinicalData", "patient_meta_concat.csv")
    # Update selected columns to use "Unnamed: 0" as the identifier
    selected_columns = ["Unnamed: 0", "Age", "Gender", "Diagnosis", "Path"]
    id_col = "Unnamed: 0"
    image_id_col = ""
    image_folder = "FundusImages"
    target_column = "Diagnosis"
    destination_path = os.path.join(dataset_path, "split")

    combine_metadatas(dataset_path, selected_columns, image_folder, metadata_path)

    metadata = preprocess_metadata(dataset_path, metadata_path, selected_columns, image_id_col, image_folder, target_column)
    metadata = process_sex(metadata, "Gender")
    metadata = process_age(metadata, "Age")
    print("Final Columns:", metadata.columns.values)
    train_split, other_split = split_train_test(metadata, id_col, test_size=0.3)
    val_split, test_split = split_train_test(other_split, id_col, test_size=0.5)
    save_split(train_split, destination_path, "train.csv", config_path, "PAPILA", "train_meta_path")
    save_split(val_split, destination_path, "val.csv", config_path, "PAPILA", "val_meta_path")
    save_split(test_split, destination_path, "test.csv", config_path, "PAPILA", "test_meta_path")
    print("PAPILA preprocessing complete.")

if __name__ == "__main__":
    preprocess_PAPILA()
```

## Covid

```python
import os
import sys
import json
import pandas as pd
from os.path import dirname as up

# Get the directory where this script resides
current_dir = os.path.dirname(os.path.abspath(__file__))

# The preprocess_helper.py file is located in the parent directory (i.e. in the preprocessing folder)
preprocessing_dir = os.path.abspath(os.path.join(current_dir, os.pardir))
if preprocessing_dir not in sys.path:
    sys.path.insert(0, preprocessing_dir)

from preprocess_helper import (
    preprocess_metadata,
    process_age,
    process_sex,
    split_train_test,
    save_split,
)

def preprocess_COVID():
    # Compute the project root directory relative to this script.
    # Given the structure:
    # /content/FedMedICLFolder/FedMedICL/preprocessing/preprocessed_dataset/COVID.py
    # The project root is at: /content/FedMedICLFolder/FedMedICL
    project_root = os.path.abspath(os.path.join(current_dir, os.pardir, os.pardir))
    config_path = os.path.join(project_root, "configs", "datasets.json")

    with open(config_path, "r") as f:
        dataset_config = json.load(f)

    dataset_path = dataset_config["COVID"]["path_to_dataset"]
    metadata_path = os.path.join(dataset_path, "metadata.csv")
    selected_columns = ["Path", "sex", "age", "view", "finding", "patientid"]

    id_col = "PATIENT"
    image_id_col = "filename"
    image_folder = "images"
    target_column = "finding"
    destination_path = os.path.join(dataset_path, "split")

    metadata = preprocess_metadata(
        dataset_path,
        metadata_path,
        selected_columns,
        image_id_col,
        image_folder,
        target_column,
        remove_missing_images=False,
        target_to_categorical=False,
        add_extention=False,
    )

    metadata = process_sex(metadata, "sex")
    metadata = process_age(metadata, "age")

    # Only select patients with COVID-19
    metadata = metadata[metadata["Target"] == "Pneumonia/Viral/COVID-19"]

    # Assign COVID as label 2, given that it will be merged with CheXpert which has labels 0 and 1
    metadata.Target = 2

    # Only keep samples with x-ray images of the views: ["PA", "AP", "AP Supine"]
    metadata = metadata[metadata["view"].isin(["PA", "AP", "AP Supine"])]

    # Rename columns to match CheXpert's naming conventions
    metadata = metadata.rename(columns={"view": "AP/PA", "patientid": "PATIENT"})
    metadata["PRIMARY_RACE"] = "Unknown"

    # Use CheXpert column order
    CheXpert_columns_order = [
        "Path",
        "Sex",
        "AP/PA",
        "Target",
        "Age_multi",
        "Age_binary",
        "PATIENT",
        "PRIMARY_RACE",
    ]
    metadata = metadata[CheXpert_columns_order]

    print("Final Columns:", metadata.columns.values)
    train_split, other_split = split_train_test(metadata, id_col, test_size=0.2)
    val_split, test_split = split_train_test(other_split, id_col, test_size=0.5)
    save_split(train_split, destination_path, "train.csv", config_path, "COVID", "train_meta_path")
    save_split(val_split, destination_path, "val.csv", config_path, "COVID", "val_meta_path")
    save_split(test_split, destination_path, "test.csv", config_path, "COVID", "test_meta_path")

if __name__ == "__main__":
    preprocess_COVID()
```

```python
import os
import sys
import torch
from PIL import Image

# Modification: Append the root directory of the repository to sys.path
sys.path.append('/content/FedMedICLFolder/FedMedICL')

# Now import BaseDataset from the dataset package
from dataset.BaseDataset import BaseDataset

# If you need AttributeGroup from the util folder, adjust the import as follows:
from dataset.util.AttributeGroup import AttributeGroup  # Assuming util contains AttributeGroup.py

class Dataset2D(BaseDataset):
    """
    A custom dataset class for handling 2D image datasets with sensitive attributes.
    """
    def __init__(
        self,
        dataframe,
        path_to_dataset,
        sensitive_names,
        tasks_sensitive_name,
        transform,
    ):
        super(Dataset2D, self).__init__(
            dataframe, path_to_dataset, sensitive_names, tasks_sensitive_name, transform
        )

    def __getitem__(self, idx):
        """
        Retrieves the image, its corresponding target label, and sensitive attributes.
        """
        item = self.dataframe.iloc[idx]
        image_path = os.path.join(self.path_to_dataset, item["Path"])
        img = Image.open(image_path).convert("RGB")
        img = self.transform(img)

        # Convert the target label from the CSV to a numeric value and then to a torch tensor.
        target_val = item["Target"]

        # Define a mapping from string phototypes to numeric labels.
        mapping = {"I & II": 0, "III": 1, "IV": 2, "V": 3}

        # Use default 0 if target_val not found in mapping.
        if isinstance(target_val, str):
            target_numeric = mapping.get(target_val, 0)
        else:
            target_numeric = target_val

        # Convert target to a torch tensor.
        target = torch.tensor(target_numeric, dtype=torch.long)

        # Get sensitive attribute/group info
        sensitive = self.get_item_group(item)

        return img, target, sensitive

    def get_labels(self):
        return self.dataframe["Target"]
```
